#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
fig, ax1 = plt.subplots(constrained_layout=True)

color = 'tab:red'
ax1.set_xlabel('x label')
ax1.set_ylabel('linear', color=color)
ax1.plot(x, x**2, color=color, label='quadratic')
ax1.tick_params(axis='y', labelcolor=color)
#ax1.set_ylim(0, 10000)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('log', color=color)  # we already handled the x-label with ax1
ax2.plot(x, np.exp(x), color=color, label='exponential')
ax2.set_yscale("log")
ax2.tick_params(axis='y', labelcolor=color)
#ax2.set_ylim(0, 10**50)

fig.tight_layout()  # otherwise the right y-label is slightly clipped

#legend documentatio:
#https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

fig.legend(bbox_to_anchor=(0.13,0.92),loc="upper left")
#fig.legend(bbox_to_anchor=(0.13,0.92),loc="upper left",frameon=False) #remove box

plt.grid(True)

plt.show()
```

```python
fitbit['Work_Weekend'] = fitbit['Work_or_Weekend'].apply(lambda x: 'Week' if x==1 else 'Weekend')
fitbit
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, figsize = (12,4))
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fitbit_month = fitbit.groupby(['Months']).agg({'Steps':'median'}).reset_index()
ax1.plot(fitbit_month['Months'],fitbit_month['Steps'],'bo',markersize = 3,color='purple',label='Median Steps per Month')
ax1.set_xlabel('Months') # Add an x-label to the axes.
ax1.set_ylabel('Steps') # Add a y-label to the axes.
ax1.legend()  # Add a legend.


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fitbit_ww = pd.DataFrame(fitbit.groupby('Work_Weekend').agg({'Steps':'mean'}).reset_index())
ax2.plot(fitbit_ww['Work_Weekend'],fitbit_ww['Steps'],'b+',markersize = 3,color='red',label='Median Steps during the Week and Weekend')
ax2.set_xlabel('Work or Weekend')  # Add an x-label to the axes.
ax2.set_ylabel('Steps')  # Add a y-label to the axes.
ax2.legend()  # Add a legend.

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
"""
cols to plot: 
x axes = 'Steps'
y axes = 'Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary'
"""
fig, ax = plt.subplots(figsize = (20,10))  # Create a figure and an axes.
yaxes = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
for y in yaxes:
    colours=['red','blue','gold']
    shapes=['go','g+','g^']
    index = colours[yaxes.index(y)]
    index_shape = shapes[yaxes.index(y)]
    ax.plot(fitbit['Steps'],fitbit[str(y)],index_shape,color=index,label=y)

ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
plt.title('A tale of 2 subplots')
ax.legend()  # Add a legend.
plt.show()
```

```python
#using seaborn

fig, ax = plt.subplots(figsize = (20,10))  # Create a figure and an axes.
yaxes = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
for y in yaxes:
    colours=['red','blue','gold']
    index = colours[yaxes.index(y)]
    ax.scatter(fitbit['Steps'],fitbit[str(y)],color=index,label=y)

ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
plt.title('A tale of 2 subplots')
ax.legend()  # Add a legend.
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# your code here
from scipy.stats import linregress
from scipy import stats

linregress(titanic['Fare'], titanic['Age']) #gives a description, not a plot

gradient, intercept, r_value, p_value, std_err = stats.linregress(titanic['Fare'], titanic['Age'])
mn=np.min(titanic['Fare'])
mx=np.max(titanic['Fare'])
x1=np.linspace(mn,mx,5)
y1=gradient*x1+intercept
plt.plot(titanic['Fare'], titanic['Age'],'ob')
plt.plot(x1,y1,'-r')
plt.show()
```
